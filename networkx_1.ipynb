```python
import networkx as nx
```

```python
G_sample = nx.Graph()#undirected 
edges = [(0,1,4),(0,3,1),(0,4,2),(1,2,5),(2,3,3),(3,4,2),(4,5,1),(2,5,2)]
G_sample.add_weighted_edges_from(edges)
pos = nx.circular_layout(G_sample)
nx.draw(G_sample,with_labels=True,node_color='orange')
edge_weights = nx.get_edge_attributes(G_sample,'weight')
a = nx.draw_networkx_edge_labels(G_sample,pos,edge_labels=edge_weights)
```

```python
print(G_sample.degree(3))
```

```python
n = list(G_sample.neighbors(3))
print(n)
```

```python
path_2_4 = nx.shortest_path(G_sample,2,4,weight='weight')
print(path_2_4)
path_2_4_length = nx.shortest_path_length(G_sample,2,4,weight='weight')
print(path_2_4_length)
```

```python
all_pair_paths = nx.shortest_path(G_sample)
print(all_pair_paths)
```

```python
all_pair_paths = nx.shortest_path(G_sample,weight='weight')
print(all_pair_paths)
```

```python
apl = nx.average_shortest_path_length(G_sample,weight='weight')
print(apl)
```

```python
closeness = nx.closeness_centrality(G_sample,distance='weight')
print(closeness)
```

```python
betweenness = nx.betweenness_centrality(G_sample,weight='weight')
print(betweenness)
```

```python
flow_value = nx.maximum_flow_value(G_sample,0,5,capacity='weight')
print(flow_value)
flow = nx.maximum_flow(G_sample,0,5,capacity='weight')
print(flow)
```

```python
#Cut: Set of edges breaking which two nodes will be disconnected

cut = nx.minimum_cut(G_sample,0,5,capacity='weight')
print(cut)
```
